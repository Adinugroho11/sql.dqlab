# List

```python
nama_teman = ['Arie', 'Budi', "Cinta", "Donny", "Elang"]
print(nama_teman)
```

```python
daftar_acak = ['Edi', 25, 175.3, 'Kelas XII IPA 5']
print(daftar_acak)
```

```python
buah = ['Apel', 'Mangga', 'Jeruk', 'Mangga']
print(buah)
```

```python
# indexing

print(nama_teman[0])
print(nama_teman[1])
print(nama_teman[2])
```

```python
# replace value

nama_teman[0] = 'Andi' # => Masukkan 'Andi' ke list nama_teman di index 0
print(nama_teman)
```

```python
nama_teman[0] == 'Arie'
# 'Andi' == 'Arie'
```

```python
print(nama_teman)
```

```python
# slicing

print(nama_teman[0:4])
```

```python
# slicing with step

print(nama_teman[::2])
# jika start tidak di-declare maka dianggap mengambil dari index 0
# jika stop tidak di-declare maka dianggap mengambil sampai value paling akhir
# jika step tidak di-declare maka dianggap step-nya 1
```

```python
nama_teman[-1]
```

```python
# slicing with step (backward)

print(nama_teman[3::-1])

# ['Andi', 'Budi', 'Cinta', 'Donny', 'Elang']
```

```python
# panjang list
len(nama_teman)
```

```python
# menambah teman baru .append()
nama_teman.append('Ferdie')
print(nama_teman)
```

```python
nama_teman_update = nama_teman.append('Galih')
```

```python
print(nama_teman_update)
```

```python
# count (menghitung berapa kali sebuah nama muncul)

nama_teman.count('Budi')
```

```python
# index (mencari nomor kamar seseorang / value)

nama_teman.index('Andi')
```

```python
# insert (memasukkan nama teman baru di index yang diinginkan)

nama_teman.insert(0, 'Galih')
print(nama_teman)
```

```python
# pop (mengambil nama yang paling belakang)

duplicate = nama_teman.pop()
print(duplicate)
```

```python
print(nama_teman)
```

```python
# pop (mengambil nama sesuai index)

random = nama_teman.pop(1)
print(random)
```

```python
print(nama_teman)
```

```python
nama_teman.append('Budi')
print(nama_teman)
```

```python
# remove (Menghapus value tertentu)
# Jika terdapat value duplikat maka yang dihapus adalah value yang muncul pertama kali
nama_teman.remove('Budi')
print(nama_teman)
```

```python
# Sort (mengurutkan secara abjad)
nama_teman.sort()
```

```python
print(nama_teman)
```

```python
# reverse (membalik urutan)
nama_teman.reverse()
```

```python
print(nama_teman)
```

```python
# variable      = list.copy() (method)
nama_teman_copy = nama_teman.copy()
nama_teman_copy
```

```python
nama_teman_copy.clear()
```

```python
print(nama_teman_copy)
```
