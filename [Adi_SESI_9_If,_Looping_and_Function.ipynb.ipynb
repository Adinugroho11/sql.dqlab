# If-Else 3 Level

```python
kondisi_badan = input("Masukan Kondisi Badan Anda : ").lower()

if kondisi_badan.isalpha() == True:
  if kondisi_badan in ['lapar', 'capek', 'ngantuk']:
    if kondisi_badan == 'lapar':
      print('makan')
    elif kondisi_badan == 'capek':
      print('istirahat')
    elif kondisi_badan == 'ngantuk':
      print('tidur')
  else:
    print("input tidak dikenali")
else:
  print("input hanya menerima alphabet")
```

```python
# buatlah kalulator BMI
# ada 2 input
# input pertama: masukkan berat badan anda (kg)
# input kedua: masukkan tinggi anda (cm)
# buatlah variabel untuk menyimpan hasil BMI
# rumus BMI = berat(kg) / tinggi(m)^2 (tidak pakai function)
# Condition:
# BMI di bawah 18.5 = Underweight
# BMI di atas sama dengan 18.5 dan di bawah 24.9 = Healthy
# Selain itu = Overweight
```

```python
# 1. Input Data
# Menggunakan float agar bisa menerima angka desimal (misal 65.5 kg)
berat_badan = float(input("Masukkan berat badan anda (kg): "))
tinggi_badan_cm = float(input("Masukkan tinggi anda (cm): "))

# 2. Konversi tinggi dari cm ke meter
# Rumus BMI wajib menggunakan satuan meter
tinggi_badan_m = tinggi_badan_cm / 100

# 3. Menghitung BMI
# Rumus: Berat / (Tinggi Meter Pangkat 2)
bmi_score = berat_badan / (tinggi_badan_m ** 2)

# 4. Logic Condition (Tanpa Function)
if bmi_score < 18.5:
    status = "Underweight"
elif 18.5 <= bmi_score < 24.9:
    status = "Healthy"
else:
    status = "Overweight"

# 5. Menampilkan Hasil
print("--------------------------------")
print(f"Skor BMI Anda : {bmi_score:.2f}") # .2f artinya dibulatkan 2 angka di belakang koma
print(f"Kategori      : {status}")
print("--------------------------------")
```

```python
# Fungsi untuk ganjil genap
# a = tanggal
# b = []

def check(tanggal, plat_nomor):
#DATA PELANGGAR
  data = {
  "melanggar" : [],
  "tidak melanggar" : []
  }
  genap = tanggal % 2 == 0
  for plat in plat_nomor:
    if genap: #tgl genap
      if plat % 2 == 0:
        data['tidak melanggar'].append(plat)
      else:
        data['melanggar'].append(plat)
    else: #tgl ganjil
      if plat % 2 == 1:
        data['tidak melanggar'].append(plat)
      else:
        data['melanggar'].append(plat)
  return data

plat_nomor = [2345, 1222, 5467, 4557, 4322, 8766, 9022]
tanggal = int(input("Masukkan tanggal (tanpa bulan tahun) hari ini: "))
hasil = check(tanggal, plat_nomor)
print(hasil)
```

```python
# Fungsi Ganjil Genap
def check_ganjil_genap(tanggal, list_plat):
    # Siapkan wadah dictionary kosong
    hasil = {
        'melanggar': [],
        'tidak melanggar': []
    }

    for plat in list_plat:
        # Cek apakah paritas (ganjil/genap) tanggal sama dengan plat
        # Jika tanggal % 2 == plat % 2, berarti SESUAI (Tidak Melanggar)
        if tanggal % 2 == plat % 2:
            hasil['tidak melanggar'].append(plat)
        else:
            hasil['melanggar'].append(plat)

    return hasil

# --- INPUT DATA (Disesuaikan agar output sama dengan request Anda) ---
a = 27  # Contoh Tanggal Ganjil
b = [1222, 4322, 8766, 9022, 2345, 5467, 4557] # Gabungan list plat

# Panggil Fungsi
data = check_ganjil_genap(a, b)

# Print Hasil
print(data)
```

```python
import asyncio
```

```python
async def math_ops(a, operator, b):
  print("start calculation")
  await asyncio.sleep(1)
  if operator == '+':
    result = a + b
  elif operator == '-':
    result = a - b
  elif operator == '*':
    result = a * b
  elif operator == '/':
    result = a / b
  return result

async def main():
  results = await asyncio.gather(
      math_ops(3, '+', 2),
      math_ops(3, '-', 2),
      math_ops(3, '*', 2),
      math_ops(3, '/', 2),
  )
  for result in results:
    print(result)

await main()
```
