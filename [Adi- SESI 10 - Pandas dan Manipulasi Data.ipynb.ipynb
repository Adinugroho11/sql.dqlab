## Mengganti type data Sales, Discount dan Profit dari Object ke Float

```python
df['Sales'] = df['Sales'].str.replace(',', '.').astype(float)
df.info()
```

```python
df['Discount'] = df['Discount'].str.replace(',', '.').astype(float)
df['Profit'] = df['Profit'].str.replace(',', '.').astype(float)
df.info()
```

```python
df.head()
```

## Mengganti type data Order_Date dan Ship_Date dari Object ke Date

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
df.info()
```

## Mengganti format nama kolom menjadi lowercase

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns
```

## Filtering

```python
df.head()
```

```python
df[df['sales'] > df['sales'].mean()]
# SELECT * FROM df WHERE sales > AVG(sales)
```

```python
df['sales'] > df['sales'].mean()
```

```python
df_sales_disc = df[(df['sales']>df['sales'].mean()) & (df['discount']>0)]
# SELECT * FROM df WHERE sales > AVG(sales) AND discount > 0
df_sales_disc
```

```python
furniture_sales = df[df['category'].isin(['Furniture', 'Technology', 'Office Supply'])]
furniture_sales
```

## Group By

```python
group_cat_sub = df[['category', 'sub-category', 'sales']].groupby(['category', 'sub-category']).sum()
group_cat_sub
```

```python
group_cat_sub.reset_index(inplace=True)
group_cat_sub
```
